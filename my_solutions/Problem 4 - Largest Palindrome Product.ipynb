A palindromic number reads the same both ways. The largest palindrome made from the product of two 2-digit numbers is 9009 = 91*99.

Find the largest palindrome made from the product of two -digit numbers.

## My solution

```python
def check_palyndrome(number):
    number_str = str(number)
    number_str_list = list(number_str)
    start = 0
    end = len(number_str_list) - 1
    is_palyndrome = False
    while (number_str_list[start] == number_str_list[end] and start < end ):
        start +=1
        end -=1
    if start > end:
        is_palyndrome = True

    return is_palyndrome

check_palyndrome(9009), check_palyndrome(9019)
```

## Official solution n°1

Let our palindrome be P = ab with a and b the two 3-digit numbers. 
If a and b are 3-digits long then they must lie between 100 and 999 inclusive. So an initial solution to the problem might be:

```python
def reverse(n):
    reversed_num = 0
    while n > 0:
        reversed_num = 10*reversed_num + n%10
        n = n//10
    return reversed_num

reverse(1912)
```

```python
def isPalindrome(n):
    return n == reverse(n)

isPalindrome(9009), isPalindrome(9019), 
```

```python
largestPalindrome = 0
a = 100
while a <= 999:
    b = 100
    while b <= 999:
        if isPalindrome(a*b) and a*b > largestPalindrome:
            largestPalindrome = a*b
        b = b+1
    a = a+1
largestPalindrome
```

## Official solution n°2

Solution n°1 is fast enough for this case but could be improved. 

For starters, the current method checks many numbers multiple times. 
For example thenumber 69696 is checked both when a=132 and b=528 and when a=528 and b=132. 
To stop checking numbers like this we can assume a ≤ b, roughly halving the number of calculations needed.

```python
largestPalindrome = 0
a = 100
while a <= 999:
    b = a
    while b <= 999:
        if isPalindrome(a*b) and a*b > largestPalindrome:
            largestPalindrome = a*b
        b = b+1
    a = a+1
largestPalindrome
```

## Official solution n°3

Next we should consider counting downwards from 999 instead of counting upwards from 100. This makes the program far more likely to find a large palindrome earlier, and we can quite easily stop checking a and b that would be too small to improve upon the largest palindrome found so far.


```python
largestPalindrome = 0
a = 999
while a >= 100:
    b = 999
    while b >= a:
        if  a*b <= largestPalindrome:
            break
        if isPalindrome(a*b):
            largestPalindrome = a*b
        b = b-1
    a = a-1
largestPalindrome
```

This is fast but can be sped up further with some analysis. Consider the digits of P.
let them be x, y and z. P must be at least 6 digits long since the palindrome 111111 = 143×777 – the product of two 3-digit integers. 
Since P is palindromic:
- P=100000x + 10000y + 1000z + 100z + 10y + x
- P=100001x + 10010y + 1100z
- P=11*(9091x + 910y + 100z)

Since 11 is prime, at least one of the integers a or b must have a factor of 11.
So if a is not divisible by 11 then we know b must be. Using this information we can determine what values of b we check depending on a:

```python
largestPalindrome = 0
a = 999
while a >= 100:
    if a % 11 == 0:
        b = 999
        db = 1
    else:
        b = 990 # //The largest number less than or equal 999 //and divisible by 11
        db = 11
    while b >= a:
        if a*b <= largestPalindrome:
            break
        if isPalindrome(a*b):
            largestPalindrome = a*b
        b = b-db
    a = a-1
largestPalindrome
```
